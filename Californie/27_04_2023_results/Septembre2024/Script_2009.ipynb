```python
def process_data2(lat1, lat2, lon1,lon2,U1, V1, C1,dt1,U2,V2,C2,dt2):
    # Trouver les éléments communs entre les deux listes de latitude avec les indices
    common, id1, id2 = np.intersect1d(lat1, lat2, return_indices=True)
    
    # Extraire les valeurs des tableaux correspondants aux indices trouvés
    U_common1 = U1[id1]
    V_common1 = V1[id1]
    C_common1 = C1[id1]
    lat_common1 = lat1[id1]
    lon_common1 = lon1[id1]
    dt_common1=dt1[id1]

    
    U_common2 = U2[id2]
    V_common2 = V2[id2]
    C_common2 = C2[id2]
    lat_common2 = lat2[id2]
    lon_common2 = lon2[id2]
    dt_common2=dt2[id2]
    
    dU=U_common2-U_common1
    dV=V_common2-V_common1
    dC=C_common2-C_common1
    d_dt=dt_common2-dt_common2
    ecart_relatif_dt = (np.abs(dt_common1 - dt_common2) / np.average([dt_common1, dt_common2])) * 100
    
    return dU, dV,dC,d_dt,lat_common1,lon_common1,ecart_relatif_dt
 
```
